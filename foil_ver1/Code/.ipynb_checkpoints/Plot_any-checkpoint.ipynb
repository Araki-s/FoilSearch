```python
import os
import matplotlib.pyplot as plt
path = "../reference/Li_backscattering/"
files = os.listdir(path)



for i in (files):
    x = []
    y = []
    print(i)
    f = open(path + i)
    line = f.readline()
    while line:
        line = line.rstrip()
        s = line.split("\t")
        x.append(float(s[0]))
        y.append(float(s[1]))
#         print(float(s[0]),float(s[1]))
        line = f.readline()
    f.close
    
    ii = i.split(".")[0]
    if i == "ENDF-8.txt":
        plt.plot(x,y, label = "            ",color = [0,173/255,74/255],marker = ".")
    elif i == "JEFF-3.3.txt":
        plt.plot(x,y, label = "            ",color = [33/255,108/255,177/255])
    elif i == "JENDL-4.0.txt":
        plt.plot(x,y, label = "         ",color = [106/255,39/255,156/255])
    else:
        plt.plot(x,y, label = "    ")
    
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0, fontsize=10,frameon=False)
plt.yscale('log')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.tick_params(bottom=False,
               left=False,
               right=False,
               top=False)
# plt.show()
plt.xlim(0,180)
plt.plot(0.001,10)
plt.savefig("300_dpi_scatter.png", format="png", dpi=600)
```
